```python
def calc_proba(churn : list) -> list:
    #[0,1,0,1,1]
    zero_count = churn.count(0)
    one_count = churn.count(1)
    total = len(churn)
    proba = []
    proba.append(zero_count/total)
    proba.append(one_count/total)

    return proba;
    
```

```python
test = [0,1,0,1,1]
print(calc_proba(test))
```
