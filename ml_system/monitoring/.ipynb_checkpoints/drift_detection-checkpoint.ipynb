```python
from ml_db.mongo_client import prediction_logs_collection
N = 5
log_data = prediction_logs_collection.find()

reference_data_cursor = prediction_logs_collection.find().sort("timestamp", 1).limit(N)
current_data_cursor = prediction_logs_collection.find().sort("timestamp", -1).limit(N)

reference_data = list(reference_data_cursor)
current_data   = list(current_data_cursor)

Numeric_feat = ["tenure", "MonthlyCharges"]
Cat_feat = ["Contract", "InternetService"]
full_feat = ["tenure", "MonthlyCharges", "Contract", "InternetService"]

feature_data = {}

for i in full_feat:
    feature_data[i] = {
        "reference_values":[
            p["features"][i]
            for p in reference_data
            if "features" in p and i in p["features"]
            ],
        "current_values": [
            p["features"][i]
            for p in current_data
            if "features" in p and i in p["features"]]
    }

numeric_distribution = {}
cat_distribution = {}

for i in Numeric_feat:
    if(i == 'tenure'):
        numeric_distribution[i] ={
            "0-20":{
                "reference%":(len([
                    p
                    for p in feature_data[i]['reference_values']
                    if 0 <= p <20
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(len([
                    p
                    for p in feature_data[i]['current_values']
                    if 0 <= p <20
                    ])/len(feature_data[i]["current_values"])) 
            },
            "20-40":{
                "reference%":(len([
                    p
                    for p in feature_data[i]['reference_values']
                    if 20 <= p <40             
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(len([
                    p
                    for p in feature_data[i]['current_values']
                    if 20 <= p <40             
                    ])/len(feature_data[i]["current_values"])) 
             
            },
            "40-60":{
                "reference%":(len([
                    p
                    for p in feature_data[i]['reference_values']
                    if 40 <= p <60             
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(len([
                    p
                    for p in feature_data[i]['current_values']
                    if 40 <= p <60             
                    ])/len(feature_data[i]["current_values"])) 
            },
            "60-80":{
                "reference%":(len([
                    p
                    for p in feature_data[i]['reference_values']
                    if 60 <= p <80             
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(len([
                    p
                    for p in feature_data[i]['current_values']
                    if 60 <= p <80             
                    ])/len(feature_data[i]["current_values"])) 
            },
            "80-100":{
               "reference%":(len([
                    p
                    for p in feature_data[i]['reference_values']
                    if 80 <= p <100             
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(len([
                    p
                    for p in feature_data[i]['current_values']
                    if 80 <= p <100             
                    ])/len(feature_data[i]["current_values"])) 
            },     
        }
    else:
        numeric_distribution[i] ={
            "18–30":{
                "reference%":(len([
                    p
                    for p in feature_data[i]['reference_values']
                    if 18 <= p <30             
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(len([
                    p
                    for p in feature_data[i]['current_values']
                    if 18 <= p <30            
                    ])/len(feature_data[i]["current_values"])) 
            },
            "30–50":{
                "reference%":(len([
                    p
                    for p in feature_data[i]['reference_values']
                    if 30 <= p <50             
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(len([
                    p
                    for p in feature_data[i]['current_values']
                    if 30 <= p <50             
                    ])/len(feature_data[i]["current_values"])) 
             
            },
            "50–70":{
                "reference%":(len([
                    p
                    for p in feature_data[i]['reference_values']
                    if 50 <= p <70             
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(len([
                    p
                    for p in feature_data[i]['current_values']
                    if 50 <= p <70             
                    ])/len(feature_data[i]["current_values"])) 
            },
            "70–90":{
                "reference%":(len([
                    p
                    for p in feature_data[i]['reference_values']
                    if 70 <= p <90             
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(len([
                    p
                    for p in feature_data[i]['current_values']
                    if 70 <= p <90             
                    ])/len(feature_data[i]["current_values"])) 
            },
            "90–120":{
               "reference%":(len([
                    p
                    for p in feature_data[i]['reference_values']
                    if 90 <= p <120             
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(len([
                    p
                    for p in feature_data[i]['current_values']
                    if 90 <= p <120             
                    ])/len(feature_data[i]["current_values"])) 
            },     
        }

for i in Cat_feat:
    if(i == "Contract"):
        cat_distribution[i] = {
                "Month-to-month":{
                    "reference%":(
                        0 if len(feature_data[i]["reference_values"])==0 
                        else
                            len([
                                p
                                for p in feature_data[i]["reference_values"]
                                if p == 'Month-to-month'
                            ])/len(feature_data[i]["reference_values"])),
                    "current%":(
                        0 if len(feature_data[i]["current_values"]) == 0
                        else
                            len([
                            p
                            for p in feature_data[i]["current_values"]
                            if p == 'Month-to-month'
                        ])/len(feature_data[i]["current_values"])),
            },
                
                
            
            "One year":{
                 "reference%":(
                     0 if len(feature_data[i]["reference_values"])==0 
                    else
                        len([
                            p
                            for p in feature_data[i]["reference_values"]
                            if p == 'One year'
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(
                    0 if len(feature_data[i]["current_values"]) == 0
                    else
                        len([
                        p
                        for p in feature_data[i]["current_values"]
                        if p == 'One year'
                    ])/len(feature_data[i]["current_values"]))

            },
            "Two year":{
                 "reference%":(
                     0 if len(feature_data[i]["reference_values"])==0 
                    else
                        len([
                        p
                        for p in feature_data[i]["reference_values"]
                        if p == 'Two year'
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(
                    0 if len(feature_data[i]["current_values"]) == 0
                    else
                        len([
                        p
                        for p in feature_data[i]["current_values"]
                        if p == 'Two year'
                    ])/len(feature_data[i]["current_values"]))
                
            }
        }
    else:
        cat_distribution[i] = {
                "No":{
                    "reference%":(
                        0 if len(feature_data[i]["reference_values"])==0 
                        else
                            len([
                                p
                                for p in feature_data[i]["reference_values"]
                                if p == 'No'
                            ])/len(feature_data[i]["reference_values"])),
                    "current%":(
                        0 if len(feature_data[i]["current_values"]) == 0
                        else
                            len([
                            p
                            for p in feature_data[i]["current_values"]
                            if p == 'No'
                        ])/len(feature_data[i]["current_values"])),
            },
                
                
            
            "Fiber optic":{
                 "reference%":(
                     0 if len(feature_data[i]["reference_values"])==0 
                    else
                        len([
                            p
                            for p in feature_data[i]["reference_values"]
                            if p == 'Fiber optic'
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(
                    0 if len(feature_data[i]["current_values"]) == 0
                    else
                        len([
                        p
                        for p in feature_data[i]["current_values"]
                        if p == 'Fiber optic'
                    ])/len(feature_data[i]["current_values"]))

            },
            "DSL":{
                 "reference%":(
                     0 if len(feature_data[i]["reference_values"])==0 
                    else
                        len([
                        p
                        for p in feature_data[i]["reference_values"]
                        if p == 'DSL'
                    ])/len(feature_data[i]["reference_values"])),
                "current%":(
                    0 if len(feature_data[i]["current_values"]) == 0
                    else
                        len([
                        p
                        for p in feature_data[i]["current_values"]
                        if p == 'DSL'
                    ])/len(feature_data[i]["current_values"]))
                
            }
        }

```

```python
from monitoring.psi import calculate_psi, classify_psi
from monitoring.drift_utils import normalize_distribution_for_psi, max_psi, num_features_drifted, num_features_major_drift, overall_status

feature_drift={}
for feature_name, dist in cat_distribution.items():
    psi_ready = normalize_distribution_for_psi(dist)
    psi_value = calculate_psi(psi_ready)

    feature_drift[feature_name] = {
        "psi": psi_value,
        "status": classify_psi(psi_value)
    }

for feature_name, dist in numeric_distribution.items():
    psi_ready = normalize_distribution_for_psi(dist)
    psi_value = calculate_psi(psi_ready)

    feature_drift[feature_name] = {
        "psi": psi_value,
        "status": classify_psi(psi_value)
    }
    
feature_drift["summary"]={
    "max_psi": max_psi(feature_drift),
    "num_features_drifted": num_features_drifted(feature_drift),
    "num_features_major_drift": num_features_major_drift(feature_drift),
    "overall_status": overall_status(feature_drift)
}


print(feature_drift)
```

```python
from datetime import datetime
from ml_db.mongo_client import feature_drift_reports_collection
report = {
      "drift_type": "feature_drift",
      "model_version": "v1",
    
      "window": {
        "type": "time_based",
        "size": N
      },
    
      "feature_metrics": feature_drift,
    
      "summary": feature_drift["summary"],
      "generated_at": datetime.utcnow().isoformat()
}
feature_drift_reports_collection.insert_one(report)
```
