```python
# inference/predict.py

import os
import joblib
import pandas as pd
from datetime import datetime

from ml_logging.prediction_logger import log_prediction

MODEL_VERSION = "v1"

def get_base_dir():
    """
    Finds project root by locating the 'model' directory.
    """
    current = os.getcwd()

    while True:
        if os.path.isdir(os.path.join(current, "model")):
            return current

        parent = os.path.dirname(current)
        if parent == current:
            raise RuntimeError("Could not find project root containing 'model/'")

        current = parent


BASE_DIR = get_base_dir()

MODEL_PATH = os.path.join(BASE_DIR, "model", "model_v1.pkl")
PREPROCESSOR_PATH = os.path.join(BASE_DIR, "model", "preprocessor.pkl")

if not os.path.exists(MODEL_PATH):
    raise FileNotFoundError(f"Model not found at {MODEL_PATH}")

if not os.path.exists(PREPROCESSOR_PATH):
    raise FileNotFoundError(f"Preprocessor not found at {PREPROCESSOR_PATH}")

model = joblib.load(MODEL_PATH)
preprocessor = joblib.load(PREPROCESSOR_PATH)

def predict(input_features: dict) -> float:
    """
    input_features: dict
        Raw feature dictionary (same schema as training)

    returns: float
        Model prediction
    """
    key_to_remove = "customer_id"
    if key_to_remove in input_features:
        customer_id = input_features.pop("customer_id")
        
    X = pd.DataFrame([input_features])
    X_transformed = preprocessor.transform(X)

    prediction = float(model.predict(X_transformed)[0])
    prediction_percentage =[]
    probas = model.predict_proba(X_transformed)[0]
    prediction_percentage.append(float(probas[1]))

    log_prediction(
        customer_id=customer_id,
        model_version=MODEL_VERSION,
        features=input_features,
        prediction=prediction,
        prediction_percentage = prediction_percentage
    )

    return prediction

if __name__ == "__main__":
    sample_input = {
        "customer_id" : "C400",
        "tenure": 10,
        "MonthlyCharges": 70.5,
        "TotalCharges": 85.0,
        "Contract": "Month-to-month",
        "PaymentMethod": "Electronic check",
        "InternetService": "Fiber optic",
        "num_support_tickets": 21
    }

    pred = predict(sample_input)
    print("Prediction:", pred)
```
